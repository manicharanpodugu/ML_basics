```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

ds=pd.read_csv('canada_PCI.csv')
ds
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("income(USD)")
plt.plot(ds.year,ds.income,color="red",marker='+')
```

```python
reg=linear_model.LinearRegression()
reg.fit(ds[['year']],ds.income)
```

```python
plt.plot(ds.year,reg.predict(ds[['year']]),color='blue')
```
