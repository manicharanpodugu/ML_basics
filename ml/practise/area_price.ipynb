```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("homeprice.csv")
df["price"]/=10
```

```python
%matplotlib inline
plt.xlabel("area(sq ft)")
plt.ylabel("price(USD)")
plt.scatter(df.area,df.price,color="red",marker='+')
```
